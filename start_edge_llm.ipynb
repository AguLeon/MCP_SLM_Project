```python
from chi import container, context, lease
import os
import chi

context.version = "1.0" 
context.choose_project()
context.choose_site(default="CHI@Edge")
```

```python
# Use the lease name that was specified 
l = lease.get_lease(f"MCPWorld_Edge_1") 
l.show()
```

```python
username = os.getenv('USER') # exp resources will have this suffix
c = container.Container(
    name = f"mcpworld-node-serve-edge".replace('_', '-'),
    reservation_id = l.device_reservations[0]["id"],
    image_ref = "docker.io/ollama/ollama:latest", 
    runtime="nvidia",
    exposed_ports = [8888, 11434, 22],
)
c.submit(idempotent=True)
```

```python
c.associate_floating_ip()
```

```python
model_name = "" # Add the model you wanna run in the edge device
```

```python
stdout, code = c.execute(f"ollama run {model_name}")
print(stdout)
```
